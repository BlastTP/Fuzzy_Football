### Przygotowanie danych (funkcje pomocnicze)

```python
def filter_and_map_columns(df, general_columns, statistics_columns_raw, mappings, statistics_columns_filtered):
    df = df[general_columns + statistics_columns_raw]
    df = df.rename(columns=mappings)
    df = df[general_columns + statistics_columns_filtered]
    return df

def drop_duplicated_columns(df):
    return df.loc[:, ~df.columns.duplicated()]

def prepare_points_and_outcome(df):
    points = lambda home_score, away_score: 3 if home_score > away_score else 1 if home_score == away_score else 0
    df['home_team_points'] = df.apply(lambda row: points(row['home_team_score'], row['away_team_score']), axis=1)
    df['away_team_points'] = df.apply(lambda row: points(row['away_team_score'], row['home_team_score']), axis=1)
    
    outcome = lambda home_score, away_score: 'win' if home_score > away_score else 'draw' if home_score == away_score else 'loss'
    df['match_outcome'] = df.apply(lambda row: outcome(row['home_team_score'], row['away_team_score']), axis=1)
    df['match_outcome_encoded'] = df['match_outcome'].map({'win': 1, 'draw': 0, 'loss': -1})
    return df
    
def prepare_momentum(df):
    home_momentum_sum = lambda data: sum(item['value'] for item in data if item['value'] > 0) if data else 0
    away_momentum_sum = lambda data: sum(abs(item['value']) for item in data if item['value'] < 0) if data else 0
    home_momentum_count = lambda data: len([item for item in data if item['value'] > 0]) if data else 0
    away_momentum_count = lambda data: len([item for item in data if item['value'] < 0]) if data else 0
    home_momentum_avg = lambda data: home_momentum_sum(data) / home_momentum_count(data) if home_momentum_count(data) > 0 else 0
    away_momentum_avg = lambda data: away_momentum_sum(data) / away_momentum_count(data) if away_momentum_count(data) > 0 else 0
    
    df['home_team_momentum_sum'] = df['momentum_graphPoints'].apply(home_momentum_sum)
    df['away_team_momentum_sum'] = df['momentum_graphPoints'].apply(away_momentum_sum)
    df['home_team_momentum_count'] = df['momentum_graphPoints'].apply(home_momentum_count)
    df['away_team_momentum_count'] = df['momentum_graphPoints'].apply(away_momentum_count)
    df['home_team_momentum_avg'] = df['momentum_graphPoints'].apply(home_momentum_avg)
    df['away_team_momentum_avg'] = df['momentum_graphPoints'].apply(away_momentum_avg)
    
    df.drop(columns=['momentum_graphPoints'], inplace=True)
    return df

def prepare_match_date(df):
    df['match_date'] = pd.to_datetime(
        df['match_day_str'] + ' ' + df['venue_time'],
        format='%A %B %d, %Y %H:%M'
    )
    return df

def calculate_past_stats_combined(df, exclude_columns=['tournament_round_number', 'team', 'team_formation', 'match_date', 'season'],
                                  past_matches=[1, 3, 5, 7, 9, 11, 13, 15]):
    home_data = df[[col for col in df.columns if col.startswith('home_') or col in exclude_columns]].copy()
    away_data = df[[col for col in df.columns if col.startswith('away_') or col in exclude_columns]].copy()

    home_data.columns = [col.replace('home_', '') for col in home_data.columns]
    away_data.columns = [col.replace('away_', '') for col in away_data.columns]

    home_data['is_home'] = 1
    away_data['is_home'] = 0

    combined_data = pd.concat([home_data, away_data]).sort_values(by=['team', 'match_date', 'season']).reset_index(drop=True)

    stat_columns = [col for col in combined_data.columns if col not in exclude_columns + ['team', 'match_date', 'season', 'is_home']]

    for n in past_matches:
        for col in stat_columns:
            sum_col = f"{col}_sum_last_{n}_matches"
            avg_col = f"{col}_avg_last_{n}_matches"

            combined_data[sum_col] = (
                combined_data.groupby('team')[col]
                .shift(1)
                .rolling(n, min_periods=n)
                .sum()
            )
            combined_data[avg_col] = (
                combined_data.groupby('team')[col]
                .shift(1)
                .rolling(n, min_periods=n)
                .mean()
            )
    
    home_stats = combined_data[combined_data['is_home'] == 1][['team', 'match_date', 'season'] + 
        [col for col in combined_data.columns if col not in ['team', 'match_date', 'season', 'is_home']]]
    home_stats.columns = [f"home_{col}" if col not in ['team', 'match_date', 'season'] else col for col in home_stats.columns]
    
    merged_df = df[['season', 'tournament_round_number', 'match_day_str', 'venue_time', 'home_team', 'away_team', 'match_date', 'match_outcome', 'match_outcome_encoded']].merge(
        home_stats, 
        left_on=['home_team', 'match_date', 'season'], 
        right_on=['team', 'match_date', 'season'], 
        how='left'
    ).drop(columns=['team', 'home_tournament_round_number'])
    
    away_stats = combined_data[combined_data['is_home'] == 0][['team', 'match_date', 'season'] + 
        [col for col in combined_data.columns if col not in ['team', 'match_date', 'season', 'is_home']]]
    away_stats.columns = [f"away_{col}" if col not in ['team', 'match_date', 'season'] else col for col in away_stats.columns]
    
    merged_df = merged_df.merge(
        away_stats, 
        left_on=['away_team', 'match_date', 'season'], 
        right_on=['team', 'match_date', 'season'], 
        how='left'
    ).drop(columns=['team', 'away_tournament_round_number'])
    
    merged_df = merged_df.sort_values(by=['match_date', 'season']).reset_index(drop=True)
    return merged_df

def fill_missing_promotion_data(df, past_matches=[1, 3, 5, 7, 9, 11, 13, 15]):
    first_season = df['season'].min()
    filtered_data = df[df['season'] > first_season]
    unique_seasons = sorted(filtered_data['season'].unique().tolist())
    
    for season in unique_seasons:
        current_season_data = df[df['season'] == season]
        
        if current_season_data.loc[current_season_data.isna().any(axis=1)].empty:
            continue
        
        past_season_data = df[df['season'] == season - 1]
        
        current_teams = set(current_season_data['home_team']).union(set(current_season_data['away_team']))
        past_teams = set(past_season_data['home_team']).union(set(past_season_data['away_team']))
        promoted_teams = current_teams - past_teams
        relegated_teams = past_teams - current_teams
        
        mapping = {promoted: relegated for promoted, relegated in zip(sorted(promoted_teams), sorted(relegated_teams))}
        
        for promoted_team, relegated_team in mapping.items():
            for n in past_matches:
                stat_columns = [
                    col for col in current_season_data.columns 
                    if col.endswith(f'_sum_last_{n}_matches') or col.endswith(f'_avg_last_{n}_matches')
                ]
                
                relegated_stats = past_season_data[
                    (past_season_data['home_team'] == relegated_team) | (past_season_data['away_team'] == relegated_team)
                ][stat_columns].tail(n)
                
                promoted_rows = current_season_data[(current_season_data['home_team'] == promoted_team) | (current_season_data['away_team'] == promoted_team)].head(n)
                
                for idx, (promoted_idx, relegated_row) in enumerate(zip(promoted_rows.index, relegated_stats.itertuples(index=False))):
                    df.loc[promoted_idx, stat_columns] = list(relegated_row)
                    
    return df

def calculate_feature_differences(df):
    home_features = [col for col in df.columns if col.startswith('home') and 'last' in col]
    away_features = [col for col in df.columns if col.startswith('away') and 'last' in col]
    
    common_names = [col.replace('home_', '').replace('away_', '') for col in home_features]
    
    assert len(home_features) == len(away_features), 'Different number of features for home and away teams'
    assert len(common_names) == len(set(common_names)), 'Duplicated feature names'
    
    for col in common_names:
        df[f'diff_{col}'] = df[f'home_{col}'] - df[f'away_{col}']
    
    return df

def add_bukmacher_data(df, df_buk):
    df = df.merge(
        df_buk,
        left_on=['season', 'home_team', 'away_team'],
        right_on=['season', 'HomeTeam', 'AwayTeam'],
    )
    return df.drop(columns=['HomeTeam', 'AwayTeam'])

def calculate_additional_columns(df, df_buk):
    # Points and outcome
    df = prepare_points_and_outcome(df)
    
    # Momentum
    df = prepare_momentum(df) if 'momentum_graphPoints' in df.columns else df
    
    # Match date
    df = prepare_match_date(df)
    
    # Past stats
    df = calculate_past_stats_combined(df)
    
    # Fill missing promotion data
    df = fill_missing_promotion_data(df)
    
    # Stats differences
    df = calculate_feature_differences(df)
    
    # Add bukmacher data
    # df = add_bukmacher_data(df, df_buk)
    
    return df

def prepare_data(df, df_buk, general_columns, statistics_columns_raw, mappings, statistics_filtered):
    # Filter columns
    df = filter_and_map_columns(df, general_columns, statistics_columns_raw, mappings, statistics_filtered)
    
    # Drop duplicated columns (due to problem in data)
    df = drop_duplicated_columns(df)
    
    # Calculate additional columns
    df = calculate_additional_columns(df, df_buk)
    
    # Drop rows with missing data
    df = df.dropna()
    
    return df

def prepare_train_test_data(df, test_seasons=[2023]):
    cols_to_remove = [col for col in df.columns.tolist() if 'team' in col and not 'diff' in col]
    cols_to_remove += ['tournament_round_number', 'match_day_str', 'venue_time', 'match_date', 'match_outcome']
    df = df.drop(columns=cols_to_remove)
    train_data = df[~df['season'].isin(test_seasons)]
    test_data = df[df['season'].isin(test_seasons)]
    train_data = train_data.drop(columns=['season'])
    test_data = test_data.drop(columns=['season'])
    
    return train_data, test_data
```

### Wczytywanie i przygotowanie danych

```python
# Wczytanie i połączenie danych z kilku sezonów
league = 'premier_league' # dostępne opcje: `premier_league`, `laliga`
first_season_year = 2019 # dostępne sezony od 2019 do 2023
last_season_year = 2023 # dostępne sezony od 2019 do 2023

# Wczytanie danych
dfs = []
last_cols = []
for season_year in range(first_season_year, last_season_year + 1):
    season_code = f'{str(season_year)[2:]}_{str(season_year + 1)[2:]}'
    df = pd.read_pickle(f'data/{league}/{season_code}/{league}_matches_{season_code}_with_fduk.pkl')
    if 'FT_GK_punches_away' in df.columns or 'FT_GK_punches_home' in df.columns:
        df = df.drop(columns=['FT_GK_punches_away', 'FT_GK_punches_home'])
    df['season'] = season_year
    dfs.append(df)

# Połączenie danych
df = pd.concat(dfs)
```

```python
# Przygotowanie danych bukmacherskich
df_buk = df.copy()
idx = df_buk.columns.tolist().index('HomeTeam')
df_buk = df_buk[df_buk.columns[idx:].tolist()]
df_buk = df_buk.dropna(axis=1)
idx_FTHG = df_buk.columns.tolist().index('FTHG')
idx_AR = df_buk.columns.tolist().index('AR')
df_buk = df_buk.drop(columns=df_buk.columns[idx_FTHG:idx_AR + 1])
```

```python
# Preprocessing i przygotowanie finalnych danych
df = prepare_data(df, df_buk, general_columns, statistics_columns_raw, mappings, statistics_filtered)
df.head()
```

```python
# Podział na zbiór treningowy i testowy oraz zapis do plików
train_data, test_data = prepare_train_test_data(df, test_seasons=[last_season_year])
train_data.to_pickle(f'data/{league}/train_data.pkl')
test_data.to_pickle(f'data/{league}/test_data.pkl')
```
